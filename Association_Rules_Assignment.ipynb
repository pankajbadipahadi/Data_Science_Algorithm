# **Data Preprocessing**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("/content/Online retail.xlsx")
df
```

```python
df.shape
```

```python
# Remove a NaN values and create perfect list
market=[]
for i in range (0,df.shape[0]):
  cus=[]
  for j in df.columns:
    if type(df[j][i])==str:
      cus.append(df[j][i])
  market.append(cus)
market
```

```python
# Create a single list
l=[]
for i in market:
  for j in i:
    l.append(j)
l
```

```python
# Find a most freaquent buy items
import collections
p=collections.Counter(l)
display(p.keys)
display(p.values)
```

```python
# Create a Data Frame
d={"Item_Names":p.keys(),"Values":p.values()}
df=pd.DataFrame(d).sort_values("Values",ascending=False)
df
```

# **Association Rule Mining**

```python
# Create Metrics
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(market).transform(market)
df_encoded=pd.DataFrame(te_ary,columns=te.columns_)
df_encoded
```

```python
from mlxtend.frequent_patterns import apriori

# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)

# Display the frequent itemsets
print("Frequent Itemsets: \n")
display(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules from frequent itemsets
rules = association_rules(frequent_itemsets, metric="lift")

rules = rules[(rules['confidence']>=0.1)]
rules=rules[(rules['lift']>=1.0)]


# Display the generated association rules
display(frequent_itemsets)
display(rules)
```

# **Analysis and Interpretation**

# Interpretation of metrics:

* **Support**: Indicates how frequently the itemset appears in the transactions.
* **Confidence**: Represents the probability that the consequent item(s) will be purchased given that the antecedent item(s) are purchased.
* **Lift**: Measures how much more likely the consequent item(s) are to be purchased when the antecedent item(s) are purchased, compared to their individual probabilities. A lift greater than 1 indicates a positive association.

```python
# Interpret the results and provide insights into customer purchasing behaviour based on the discovered rules.
print("Interpreting the Association Rules:")

if rules.empty:
    print("No association rules were found with the current thresholds. Consider lowering the support, confidence, or lift values.")
else:
    print("Here are some insights based on the discovered rules:")
    # You can add more specific interpretations based on the generated rules.
    # For example, you can iterate through the rules and print insights for each.
    for index, row in rules.iterrows():
        antecedents = list(row['antecedents'])
        consequents = list(row['consequents'])
        confidence = row['confidence']
        lift = row['lift']

        print(f"- Customers who bought {antecedents} are likely to also buy {consequents}.")
        print(f"  - This relationship has a confidence of {confidence:.2f} and a lift of {lift:.2f}.")
```

# **Interview Question**

**1.	What is lift and why is it important in Association rules?**

Lift is a metric used in association rule mining to measure how much more likely a consequent item(s) is to be purchased when an antecedent item(s) is purchased, compared to the base probability of purchasing the consequent item(s) alone. It indicates the strength of the association between the antecedent and consequent.

Why is it important?

* Identifies True Associations: A lift value greater than 1 suggests a positive correlation between the items, meaning the antecedent item(s) has a positive influence on the purchase of the consequent item(s).
* Filters out Spurious Rules: Lift helps to filter out rules that might have high confidence but are simply due to the high overall popularity of the consequent item(s). A high lift indicates a relationship beyond what would be expected by chance.
* Guides Business Decisions: By identifying rules with high lift, businesses can gain valuable insights into customer purchasing patterns, which can inform decisions related to product placement, promotions, and recommendations.

**2.	What is support and Confidence. How do you calculate them?**

Here are the definitions and calculations for Support and Confidence in association rule mining:

* Support: Support is an indication of how frequently the itemset appears in the dataset. It is calculated as the number of transactions containing the itemset divided by the total number of transactions.

$$ \text{Support}(A \cup B) = \frac{\text{Number of transactions containing both A and B}}{\text{Total number of transactions}} $$

* Confidence: Confidence is a measure of how often items in B appear in transactions that contain A. It is calculated as the number of transactions containing both A and B divided by the number of transactions containing A.

$$ \text{Confidence}(A \Rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)} = \frac{\text{Number of transactions containing both A and B}}{\text{Number of transactions containing A}} $$

**3.	What are some limitations or challenges of Association rules mining?**

Here are some limitations and challenges of Association Rule Mining:

* Generating too many rules: In datasets with many items, the Apriori algorithm (or similar algorithms) can generate an extremely large number of rules, many of which may not be practically useful or insightful.
* Sparsity: If the dataset is sparse (i.e., most transactions contain only a small subset of the total items), it can be challenging to find strong associations with high support.
* Determining appropriate thresholds: Choosing the right minimum support and confidence thresholds is crucial. If they are too high, you might miss interesting rules; if they are too low, you might generate too many irrelevant rules.
* Computational complexity: For very large datasets, the process of finding frequent itemsets can be computationally expensive.
* Interpreting complex rules: Rules involving many items can be difficult to interpret and translate into actionable business insights.
* Handling quantitative data: Association rule mining is primarily designed for categorical data. Handling quantitative data requires discretization, which can lead to loss of information.
* Lack of causal inference: Association rules show correlation, not causation. Just because two items are frequently bought together doesn't mean one causes the other to be purchased.
* Data quality: The quality of the results is highly dependent on the quality of the input data. Inaccurate or incomplete data can lead to misleading rules.
