Apply scaling techniques to numerical features:

1)	Standard Scaling
2)	Min-Max Scaling

```python
from sklearn.preprocessing import StandardScaler

numerical_features = ['age','fnlwgt','capital_gain','capital_loss','hours_per_week']

# Standard Scaling
ss = StandardScaler()
df_scaled_standard = df.copy() # create a copy to avoid modifying original df
df_scaled_standard[numerical_features] = ss.fit_transform(df[numerical_features])
print("DataFrame after Standard Scaling:")
display(df_scaled_standard.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

numerical_features = ['age','fnlwgt','capital_gain','capital_loss','hours_per_week']

# Min-Max Scaling
mms = MinMaxScaler()
df_scaled_minmax = df.copy() # create another copy
df_scaled_minmax[numerical_features] = mms.fit_transform(df[numerical_features])
print("\nDataFrame after Min-Max Scaling:")
display(df_scaled_minmax.head())
```

**Discuss the scenarios where each scaling technique is preferred and why**

**1) Standard Scaling**: This technique is preferred when the data is normally distributed or approximately normally distributed. It transforms the data such that the mean is 0 and the standard deviation is 1. This is beneficial for algorithms that assume a Gaussian distribution of the input features, such as:
* Linear Regression
* Logistic Regression
* Linear Discriminant Analysis (LDA)
* Principal Component Analysis (PCA)

Standard Scaling is also less affected by outliers compared to Min-Max Scaling.

**2)Min-Max Scaling**: This technique scales the data to a fixed range, usually between 0 and 1. It's useful when:

You need the features to be within a specific range (e.g., in image processing).
The algorithm you are using is not sensitive to the distribution of the data.
Algorithms that rely on distance calculations, such as:
* K-Nearest Neighbors (KNN)
* Support Vector Machines (SVM) with Radial Basis Function (RBF) kernel
* Neural Networks (often benefit from input features scaled to a specific range)

Min-Max Scaling is sensitive to outliers, as they will be scaled to the minimum or maximum value in the new range

# **Task 2 : Encoding Techniques**

**Apply One-Hot Encoding to categorical variables with less than 5 categories.**

```python
df_ohe=df.copy()
categorical_cols = df_ohe.select_dtypes(include='object').columns
low_cardinality_cols = [col for col in categorical_cols if df_ohe[col].nunique() < 5]

df_encoded = pd.get_dummies(df_ohe, columns=low_cardinality_cols, drop_first=True)

print("Original DataFrame shape:", df_ohe.shape)
print("DataFrame shape after One-Hot Encoding:", df_encoded.shape)
display(df_encoded.head())
```

**Use Label Encoding for categorical variables with more than 5 categories.**

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = df.select_dtypes(include='object').columns
high_cardinality_cols = [col for col in categorical_cols if df[col].nunique() >= 5]

le = LabelEncoder()
df_le = df.copy() # Create a copy to avoid modifying original df

for col in high_cardinality_cols:
    df_le[col] = le.fit_transform(df_le[col])

print("\nDataFrame after Label Encoding high cardinality columns:")
display(df_le.head())
```

**Discuss the pros and cons of One-Hot Encoding and Label Encoding.**

**1) One-Hot Encoding**:
* Pros : Creates a new binary column for each category, avoiding the assumption of ordinal relationships between categories.
Suitable for nominal categorical variables where the order doesn't matter.
Most machine learning algorithms can directly use one-hot encoded data.

* Cons : Can significantly increase the dimensionality of the dataset, especially with high-cardinality categorical variables (many unique categories). This can lead to the "curse of dimensionality."
Can lead to sparse data (many zero values), which can be computationally expensive for some algorithms.

**2) Label Encoding**:
* Pros : Reduces the number of features compared to One-Hot Encoding, especially for high-cardinality variables.
Can be useful for ordinal categorical variables where there is a natural order or ranking between categories.

* Cons : Introduces an artificial ordinal relationship between categories that don't have a natural order. This can mislead some machine learning algorithms into assuming a hierarchy that doesn't exist.
Not suitable for nominal categorical variables.

# **Task 3 : Feature Engineering**

**Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.**

```python
# Create 'Total_Capital_Activity' feature
df['Total_Capital_Activity'] = df['capital_gain'] - df['capital_loss']

# Create 'Age_Group' feature by discretizing age
df['Age_Group'] = pd.cut(df['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Adult', 'Middle-aged', 'Senior'])

print("\nDataFrame with new features:")
display(df.head())
```

Rationale for new features:

**1) Total_Capital_Activity**: This feature combines the capital_gain and capital_loss into a single metric. The rationale is that the net capital activity might be more informative than considering gains and losses separately. A positive value indicates a net gain, while a negative value indicates a net loss. This could potentially capture a person's overall investment behavior and its impact on income.

**2) Age_Group**: This feature categorizes the continuous age variable into discrete groups (Young, Adult, Middle-aged, Senior). The rationale is that income levels and patterns can vary significantly across different age groups. By creating these bins, we can help the model capture non-linear relationships between age and income that might not be well-represented by the raw age value. This can simplify the model and potentially improve its performance by highlighting distinct characteristics of each age group.

# **Task 4 : Feature Selection**

**Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance.**

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# Select numerical features for outlier detection
numerical_features = ['age', 'fnlwgt', 'capital_gain_log', 'hours_per_week']

# Initialize and train the Isolation Forest model
# contamination is the proportion of outliers in the data set.
iso_forest = IsolationForest(contamination=0.05, random_state=42)
iso_forest.fit(df[numerical_features])

# Predict outliers ( -1 for outliers and 1 for inliers)
outlier_predictions = iso_forest.predict(df[numerical_features])

# Create a new DataFrame without outliers
df_no_outliers = df[outlier_predictions == 1].copy()

print("Original DataFrame shape:", df.shape)
print("DataFrame shape after removing outliers:", df_no_outliers.shape)
display(df_no_outliers.head())
```

How Outliers Can Affect Model Performance

Outliers can significantly impact the performance of many machine learning models, particularly those that are sensitive to the distribution of the data or rely on distance calculations. Here's how:

**1) Skewing Statistics**: Outliers can heavily influence descriptive statistics like the mean and standard deviation, which can misrepresent the central tendency and spread of the data.

**2) Distorting Model Training**:
* Regression Models (e.g., Linear Regression): Outliers can pull the regression line away from the true underlying relationship, leading to a poor fit and inaccurate predictions.
* Distance-Based Models (e.g., K-Nearest Neighbors, Support Vector Machines): Outliers can disproportionately affect distance calculations, leading to incorrect nearest neighbors or poorly defined decision boundaries.
* Tree-Based Models (e.g., Decision Trees, Random Forests): While generally more robust to outliers than linear models, extreme outliers can still influence tree splits, especially in the initial stages.
* Clustering Algorithms (e.g., K-Means): Outliers can significantly affect the calculation of cluster centroids, leading to incorrect cluster assignments and a poor clustering structure.

**3) Inflating Error Rates**: Models trained on data with outliers may have higher error rates, as the model tries to accommodate these extreme values.

**4) Reducing Model Interpretability**: The presence of outliers can make it harder to interpret the relationships between features and the target variable.

Removing or transforming outliers can help to mitigate these issues, leading to more robust, accurate, and interpretable models.

```python
import ppscore as pps
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the PPS matrix
pps_matrix = pps.matrix(df)[['x', 'y', 'ppscore']].pivot(columns='x', index='y', values='ppscore')

print("Predictive Power Score (PPS) Matrix:")
display(pps_matrix)

# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

print("\nCorrelation Matrix:")
display(correlation_matrix)

# Visualize the PPS matrix
plt.figure(figsize=(12, 10))
sns.heatmap(pps_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title('Predictive Power Score (PPS) Matrix')
plt.show()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

**Discussion on PPS vs. Correlation Matrix**

* **Correlation Matrix**: Measures the linear relationship between numerical features. Values range from -1 to 1, where 1 indicates a perfect positive linear correlation, -1 indicates a perfect negative linear correlation, and 0 indicates no linear correlation. Correlation only captures linear relationships and is not suitable for categorical variables or non-linear relationships.

* **Predictive Power Score (PPS)**: Measures how well one feature can predict another feature. The score ranges from 0 to 1, where 0 means the predictor feature has no predictive power for the target feature, and 1 means the predictor feature can perfectly predict the target feature. PPS can capture both linear and non-linear relationships and works for both numerical and categorical variables. It is asymmetric (PPS(a, b) is not necessarily equal to PPS(b, a)).

**Comparison**:

The correlation matrix only shows relationships between numerical features, while the PPS matrix shows relationships between all pairs of features (numerical and categorical).
The PPS matrix can reveal non-linear relationships that the correlation matrix might miss.
The PPS matrix is asymmetric, reflecting the predictive power in one direction (from predictor to target), while the correlation matrix is symmetric, reflecting the linear association between two variables.
PPS is particularly useful for identifying which features are strong predictors of the target variable or other features, even if the relationship is not linear.
